```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Giả sử bạn có một DataFrame với dữ liệu việc làm
data = {
    'Khu_vuc': ['Nong_thon', 'Nong_thon', 'Thanh_pho', 'Thanh_pho'],
    'Muc_thu_nhap': [150, 200, 500, 600],
    'Ty_le_that_nghiep': [5, 4, 7, 3],
    'So_luong_lao_dong': [100, 150, 200, 250]
}

df = pd.DataFrame(data)

# Tính toán một số thống kê cơ bản
summary = df.groupby('Khu_vuc').agg({'Muc_thu_nhap': 'mean', 'Ty_le_that_nghiep': 'mean', 'So_luong_lao_dong': 'sum'})

print(summary)

# Vẽ biểu đồ thu nhập
plt.figure(figsize=(10, 5))
sns.barplot(x=summary.index, y='Muc_thu_nhap', data=summary)
plt.title('Mức thu nhập trung bình theo khu vực')
plt.xlabel('Khu vực')
plt.ylabel('Mức thu nhập (VND)')
plt.show()

# Vẽ biểu đồ tỷ lệ thất nghiệp
plt.figure(figsize=(10, 5))
sns.barplot(x=summary.index, y='Ty_le_that_nghiep', data=summary)
plt.title('Tỷ lệ thất nghiệp theo khu vực')
plt.xlabel('Khu vực')
plt.ylabel('Tỷ lệ thất nghiệp (%)')
plt.show()
```
